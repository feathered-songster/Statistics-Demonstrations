# Beyond the Arithmetic Mean

The **arithmetic mean** (or the boring one) is literally everywhre. Just look how many functions and methods we have for it. 
But as we'll see in this notebook, there are other useful means out there.

```python
import statistics
import numpy as np
import pandas as pd
from scipy import stats
```

```python
data = [3, 5, 7, 9, 11]

print("Python:", sum(data) / len(data))
print("statistics:", statistics.mean(data))
print("NumPy:", np.mean(data))
print("NumPy average:", np.average(data))
print("Pandas:", pd.Series(data).mean())
print("Pandas2:", pd.DataFrame(data).mean())
print("SciPy:", stats.tmean(data))  # stands fro trimmed mean and can take trimming partameters
print("ndarray method:", np.array(data).mean())
```

## Weighted Mean

While on the topic of boring means, let's get the **weighted mean** out of the way - you know the one where homework counts for 20%, a midterm for 30%, and a final exam for 50%. 

The weighted mean is calculated as

$$
\bar{x}_w = \frac{\sum_{i=1}^{n} w_i x_i}{\sum_{i=1}^{n} w_i}
$$

where $x_i$ are the observations and $w_i$ are their corresponding weights.


```python
scores = [85, 90, 78]  # homework, midterm, final
weights = [0.20, 0.30, 0.50]  # 20%, 30%, 50%

np.average(scores, weights=weights)
```

## Geometric Mean

The arithmetic mean is great when we are adding things up, but some quantities grow by multiplication rather than addition. When we are dealing with growth rates, percentages, or repeated changes over time, the **geometric mean**  gives a more meaningful average.

Imagine a bee population in a protected area. Over three years, the population changes by different growth factors:

- Year 1: the population doubles (×2)
- Year 2: the population grows by 50% (×1.5)
- Year 3: the population grows by 20% (×1.2)

Given that growth compounds: each year's change builds on the previous year's population. The geometric mean finds **the constant growth factor that would produce the same overall change**.

The geometric mean is calculated as:

$$
G = \sqrt[n]{x_1 \times x_2 \times ... \times x_n}
$$

where $x_i$ are the values being averaged and $n$ is the number of observations.


```python
# to manually calcumate use the 1/n exponent notation

growth_factors = [2, 1.5, 1.2]
np.prod(growth_factors) ** (1 / len(growth_factors))
```

```python
from scipy.stats import gmean

gmean(growth_factors)
```

By the way (in case you want to flex), the log of the geometric mean is just the arithmetic mean of the logs.

$$
G = \sqrt[n]{x_1 \times x_2 \times ... \times x_n}
$$

Taking the log

$$
\log(G) = \frac{\log(x_1)+\log(x_2)+...+\log(x_n)}{n}
$$

So we have 

$$
G = \exp\left(\frac{\log(x_1)+\log(x_2)+...+\log(x_n)}{n}\right)
$$

Jokes aside, this is useful to **avoid overflow and underflow during the multiplication step when working with very large or very small numbers**.

```python
np.exp(np.mean(np.log(growth_factors)))
```

## Harmonic Mean

The arithmetic mean works well when we are averaging quantities like grades or measurements, but it can be misleading when dealing with **rates**. When the values represent "how much per unit" (such as speed, population per suqare kilometer, etc.), the harmonic mean is the better choice.

Imagine driving the same distance twice:

- On the way there, you drive at **60 km/h**
- On the way back, you drive at **40 km/h**

Obviously we cannot use the arithmetic mean,as this would ignore the fact that you spend more time travelling at the slower speed. Since the distances are equal, the correct average speed is lower.

The harmonic mean is calculated as:

$$
H = \frac{n}{\frac{1}{x_1}+\frac{1}{x_2}+...+\frac{1}{x_n}}
$$

where $x_i$ are the values being averaged and $n$ is the number of observations.

For rates, the harmonic mean gives **the average rate that would produce the same total result**.

Note we can ignore the actual distance travelled here, but we do assume the distance is the same back and forth!

```python
speeds = [60, 40]
len(speeds) / np.sum(1 / np.array(speeds))
```

```python
from scipy.stats import hmean

hmean(speeds)
```

## Trimmed Mean

Sometimes the ordinary mean is too sensitive to extreme values. A single unusually large or small observation can pull the average away from what is typical.

A **trimmed mean** reduces this effect by removing a chosen percentage of the smallest and largest observations before calculating the average.

For example, in a dataset of test scores, a few unusually low or high scores might represent unusual circumstances rather than the typical performance. A trimmed mean can provide a more robust summary.

SciPy provides this through `scipy.stats.tmean()`.


```python
from scipy import stats

data=[20 , 30, 40, 100_000] 
stats.tmean(data, limits=(10, 90))
```

As an alternative those extremes may be kept but replaced with the upper limit chosen. Consider the analysis goal.

If needed, we can consider using more robust variants of the mean too, say Winsorized Mean, when we don't want the most or least extreme 5% of values to dominate our average, regardless of whether they are technically outliers.
